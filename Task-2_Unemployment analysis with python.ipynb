# Task-2: UNEMPLOYMENT ANALYSIS WITH PYTHON

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
data=pd.read_csv("Unemployment rate in India.csv")
```

```python
data.columns
```

##  Data cleaning

```python
#Removing the spaces in column names
data =data.rename(columns=lambda x: x.strip())
```

```python
data.columns
```

```python
## Checking for null values
data.isnull().sum()
```

### There are no null values

```python
### checking for missing values
data.isna().sum()
```

### There are no missing values

```python
data.info()
```

```python
### Converting the data column to datetime datatype
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.info()
```

## Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#correlation between the features of this dataset
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr())
plt.show()
```

```python
# Analyzing the unemployment rate over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=data)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate in India Over Time')
plt.xticks(rotation=45)
plt.show()
```

### From the plot, we can observe that employment rate is high during the months of May and June

```python
# Unemployment rates according to diiferent regions
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x='Estimated Unemployment Rate (%)', hue="Region.1", data=data)
plt.show()
```

### From the above plot we can observe that South region has the highest unemployment rate

```python
# Analyzing the relationship between unemployment rate and number of employed
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Estimated Employed', y='Estimated Unemployment Rate (%)', data=data)
plt.xlabel('Estimated Employed')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate vs. Number of Employed')
plt.show()
```

```python
# Analyzing the relationship between unemployment rate Labour participation rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Estimated Labour Participation Rate (%)', y='Estimated Unemployment Rate (%)', data=data)
plt.xlabel('Labour Participation Rate (%)')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate vs. Labour Participation Rate')
plt.show()
```

```python
#Line Plot of Unemployment Rate by Region:
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', hue='Region', data=data)
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate by Region Over Time')
plt.xticks(rotation=45)
plt.show()
```
